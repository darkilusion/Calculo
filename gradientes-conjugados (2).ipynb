¿Qué es el gradiente conjugado?

El método de gradiente conjugado es uno de los más usados para poder resolver grandes sistemas lineales de ecuaciones.
Este método es útil para sistemas de la siguiente forma: 

$$Ax = b$$

en donde "x" es un vector desconocido, b es un vector conocido, y A es una matriz cuadrada simetrica y definida positiva, y ademas conocida. Estos sistemas de ecuaciones aparecen en la aplicación de técnicas numéricas de gran importancia como por ejemplo los métodos de diferencias finitas o elementos finitos para resolver ecuaciones diferenciales en derivadas parciales.

Veremos un ejemplo de los gradientes conjugados en Python:

Importamos las librerias que utilizaremos.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
import itertools
```

Con la libreria numpy crearemos matrices:

```python
A = np.matrix([[3.0, 2.0], [2.0, 6.0]])
b = np.matrix([[2.0], [-8.0]])
c = 0.0
```

Obtenemos la función Cuadrática de la siguiente forma:

$$ f(x) = \frac{1}{2}x^\text{t}Ax - b^\text{t} x + c $$

$$ f'(x) = Ax - b $$

$$ Ax - b = 0 $$

Donde podemos ver que A es una matriz cuadrada, x y b son los vectores y por ultimo la c que corresponde a una constante escalar.

En esta funcion se retorna el valor de la funcion cuadratica.

```python
def f(x, A, b, c):
    return float(0.5 * x.T * A * x - b.T * x + c)
```

Luego evaluamos la funcion y la dibujamos por medio de la libreria matplotlib. Ademas de obtener el grafico de contorno de la misma forma cuadrática.

```python
#función en una cuadrática.

def bowl(A, b, c):
    fig = plt.figure(figsize=(10,8))
    qf = fig.gca(projection='3d')
    size = 20
    x1 = list(np.linspace(-6, 6, size))
    x2 = list(np.linspace(-6, 6, size))
    x1, x2 = np.meshgrid(x1, x2)
    zs = np.zeros((size, size))
    for i in range(size):
        for j in range(size):
            x = np.matrix([[x1[i,j]], [x2[i,j]]])
            zs[i,j] = f(x, A, b, c)
    qf.plot_surface(x1, x2, zs, rstride=1, cmap=cm.coolwarm, cstride=1, linewidth=0)
    fig.show()
    return x1, x2, zs

#contorno de la misma forma cuadrática.

def contoursteps(x1, x2, zs, steps=None):
    fig = plt.figure(figsize=(6,6))
    cp = plt.contour(x1, x2, zs, 10)
    plt.clabel(cp, inline=1, fontsize=10)
    if steps is not None:
        steps = np.matrix(steps)
        plt.plot(steps[:,0], steps[:,1], '-o')
    fig.show()

x1, x2, zs = bowl(A, b, c)
contoursteps(x1, x2, zs)
```
